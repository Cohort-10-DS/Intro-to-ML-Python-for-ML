### Basic Scatter Plot: Sepal_length vs. Sepal_Width

```python
plt.plot(df_x.sepal_length,
df_x.sepal_width,
ls ='', marker='o',)
plt.xlabel("length") # x-axis
plt.ylabel("width") # y-axis
```

Discussion : The .plot methods supports a lot of arguements which allow us to customize the visualization to a high degree. For example if we want to change our data plot from circle markers to x's or to increase the size of the markers. 

(Check out the documentation here to see the entire range of customizations: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html)


```python
plt.plot(df_x.sepal_length,
df_x.sepal_width,
ls ='', marker='x',markersize = 20, color = "green")
plt.xlabel("length") # x-axis
plt.ylabel("width") 
```

### Layered Scatter Plot: Sepal vs Petal

```python
plt.plot(df_x.sepal_length,
df_x.sepal_width,
ls ='', marker='o',
label='sepal_length')
plt.legend() # add legend

plt.plot(df_x.petal_length,
df_x.petal_width,
ls ='', marker='o',
label='petal_length')
plt.legend() # add legend
```

# Seaborn 

While matplotlib and Seaborn have a lot of similar functionality, Seaborn is especially good statistial data  visualizations.

Lets visualize the distributions of two variables along with their pearson correlation coefficient.

```python
sns.jointplot(x='sepal_length',
y='sepal_width',
data=df_x, size=6)
```

### Pop Quiz

How could we print the joint plot for all pairs of variables pythonically?

```python
def print_allpairs():
    for i in df_x:
        for j in df_x:
            if i !=j: 
                sns.jointplot(x=i,
                y=j,
                data=df_x, size=5)
print_allpairs()
```
